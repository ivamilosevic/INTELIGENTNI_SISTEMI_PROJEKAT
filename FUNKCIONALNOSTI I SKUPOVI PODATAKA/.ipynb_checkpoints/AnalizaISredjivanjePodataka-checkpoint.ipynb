### Provera da li su manji dataset-ovi sadržani u velikom, tj da li je najveći skup podataka zbir manjih

```python
reviews = airline_df[['airline_name', 'content']].copy()
lounges = lounge_df[['airline_name', 'content']].copy()
seats = seat_df[['airline_name', 'content']].copy()

#normalizacija i čišćenje teksta
def normalize(text):
    if pd.isna(text):
        return ""
    return (
        text.lower()
        .strip()
        .replace('\n', ' ')
        .replace('\r', ' ')
    )

#kreiranje nove kolone u svakom dataset-u sa očišćenim recenzijama
for df in [reviews, lounges, seats]:
    df['content_norm'] = df['content'].apply(normalize)

#dodavanje kolone koja će označavati koji je dataset u pitanju
reviews['source'] = 'reviews'
lounges['source'] = 'lounges'
seats['source'] = 'seats'

#spajanje svih tih podataka u jedan dataframe(kolone airline_name, content, content_norm i source)
all_content = pd.concat([reviews, lounges, seats], ignore_index=True)

#print(all_content)

#duplikati se tra\e po aviokompaniji i normalizovanom tekstu koji se odnosi na recenzije
duplicates = all_content[
    all_content.duplicated(
        subset=['airline_name', 'content_norm'],
        keep=False
    )
]

duplicates.sort_values(['airline_name', 'content_norm'])

#definisanje kako će se prikazati duplikati, koliko puta i iz kojih izvora
dup_stats = (
    duplicates
    .groupby(['airline_name', 'content_norm'])
    .source
    .nunique()
    .reset_index(name='num_sources')
)

dup_stats.sort_values('num_sources', ascending=False).head(10)

print("Ukupno zapisa:", len(all_content))
print("Duplikati:", duplicates[['airline_name', 'content_norm']].drop_duplicates().shape[0])
print("Duplikati iz više izvora:", (dup_stats['num_sources'] > 1).sum())
```

### S obzirom na rezultate može se zaključiti da manji dataset-ovi nisu sadržani u velikom, već predstavljaju drugu vrstu recenzija za specifične aspekte putovanja

### Cilj je identifikovati avio-kompanije koje se pojavljuju u sva tri dataseta (airline, lounge i seat). Pored toga, postoji veza između dataseta lounge i airport, koja se ostvaruje preko atributa airport u okviru lounge dataset-a, što omogućava analizu kvaliteta lounge-a u kontekstu određenih aerodroma.

```python
airlines_reviews = set(airline_df['airline_name'].unique())
airlines_lounges_reviews = set(lounge_df['airline_name'].unique())
airlines_seats_reviews = set(seat_df['airline_name'].unique())

matching_airlines = airlines_reviews & airlines_lounges_reviews & airlines_seats_reviews

print(sorted(matching_airlines))
print("Broj poklapanja aviokompanija u sva tri skupa podataka:", len(matching_airlines))
```

```python
lounge_airports = lounge_df['airport'].dropna().unique()
airport_names = airport_df['airport_name'].dropna().unique()

def normalize_airport(name):
    return name.lower().replace(" ", "-")

lounge_airports_norm = set(normalize_airport(a) for a in lounge_airports)
airport_names_norm = set(normalize_airport(a) for a in airport_names)

matching_airports = lounge_airports_norm & airport_names_norm
print("Broj poklapanja aerodroma u dva skupa podataka:", len(matching_airports))
print(sorted(matching_airports))
```

### Za ovaj zadatak koristiće se manji skupovi podataka u kojima se nalaze više recenzija za 20 avio-kompanija. Broj avio-kompanija se može lako proširiti, pri čemu bi pristup i metodologija ostali nepromenjeni.

```python
# Filtriranje po zajedničkim aviokompanijama
flight_counts = airline_df[airline_df['airline_name'].isin(matching_airlines)] \
                    ['airline_name'].value_counts()

lounge_counts = lounge_df[lounge_df['airline_name'].isin(matching_airlines)] \
                    ['airline_name'].value_counts()

seat_counts = seat_df[seat_df['airline_name'].isin(matching_airlines)] \
                    ['airline_name'].value_counts()


total_counts = flight_counts.add(lounge_counts, fill_value=0) \
                            .add(seat_counts, fill_value=0)

# izbor 20 aviokompanija za koje ima najviše recenzija u skupu podataka
top_20_airlines = total_counts.sort_values(ascending=False).head(20)

print(top_20_airlines)
```

```python
#  Flight dataset
existing_top_airlines_flight = [x for x in top_20_airlines.index if x in airline_df['airline_name'].values]
airline_df_top_20 = airline_df[airline_df['airline_name'].isin(existing_top_airlines_flight)].copy()
print("Flight dataset:")
print(airline_df_top_20.info())

#  Lounge dataset 
existing_top_airlines_lounge = [x for x in top_20_airlines.index if x in lounge_df['airline_name'].values]
lounge_df_top_20 = lounge_df[lounge_df['airline_name'].isin(existing_top_airlines_lounge)].copy()
print("\nLounge dataset:")
print(lounge_df_top_20.info())

#  Seat dataset 
existing_top_airlines_seat = [x for x in top_20_airlines.index if x in seat_df['airline_name'].values]
seat_df_top_20 = seat_df[seat_df['airline_name'].isin(existing_top_airlines_seat)].copy()
print("\nSeat dataset:")
print(seat_df_top_20.info())
```

### Nazive avio-kompanija izmenićemo tako da budu bez crtica i sa velikim početnim slovima.

```python
# Svi jedinstveni nazivi iz sva tri dataset-a
all_airlines = set(airline_df_top_20['airline_name'].unique()) \
             | set(lounge_df_top_20['airline_name'].unique()) \
             | set(seat_df_top_20['airline_name'].unique())

# Pretvaranje u sortiranu listu 
all_airlines_sorted = sorted(all_airlines)

print("Sve aviokompanije u top 20 dataset-ima (bez dupliranja):")
print(all_airlines_sorted)
```

```python
def format_airline_name(df):
    df['airline_name'] = (
        df['airline_name']
        .str.replace('-', ' ', regex=False)
        .str.title()
    )
    return df

airline_df_top_20 = format_airline_name(airline_df_top_20)
lounge_df_top_20 = format_airline_name(lounge_df_top_20)
seat_df_top_20 = format_airline_name(seat_df_top_20)
```

### Pretvorićemo vrednosti 1 i 0 u yes i no radi preglednosti i buduće analize

```python
airline_df_top_20['recommended'] = airline_df_top_20['recommended'].map({1: 'yes', 0: 'no'})
lounge_df_top_20['recommended'] = lounge_df_top_20['recommended'].map({1: 'yes', 0: 'no'})
seat_df_top_20['recommended'] = seat_df_top_20['recommended'].map({1: 'yes', 0: 'no'})
```

### Nedostajući podaci biće dopunjeni na sledeći način: numerički podaci će se zameniti srednjom vrednošću, dok će atributi tipa object dobiti vrednost unknown.

```python
def clean_top20_df(df, numeric_cols=None, obj_cols=None):

    df_clean = df.copy()

    # automatsko detektovanje numeričkih kolona
    if numeric_cols is None:
        numeric_cols = df_clean.select_dtypes(include=['float64','int64']).columns
    
    # automatsko detektovanje object kolona
    if obj_cols is None:
        obj_cols = df_clean.select_dtypes(include=['object']).columns
    
    # popunjavanje numeričkih kolona
    df_clean[numeric_cols] = df_clean[numeric_cols].fillna(df_clean[numeric_cols].mean()).round(1)
    
    # popunjavanje kolona tipa object 
    for col in obj_cols:
        df_clean[col] = df_clean[col].fillna("Unknown")
    
    return df_clean

airline_top20_df_filtered = clean_top20_df(airline_df_top_20)
lounge_top20_df_filtered = clean_top20_df(lounge_df_top_20)
seat_top20_df_filtered = clean_top20_df(seat_df_top_20)

print("Flight dataset (top 20) info:")
print(airline_top20_df_filtered.info())

print("\nLounge dataset (top 20) info:")
print(lounge_top20_df_filtered.info())

print("\nSeat dataset (top 20) info:")
print(seat_top20_df_filtered.info())
```
